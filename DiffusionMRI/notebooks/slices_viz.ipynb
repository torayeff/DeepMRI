```python
import nibabel as nib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.axes_grid1 import make_axes_locatable
%matplotlib inline
```

```python
pth = "/home/agajan/experiment_DiffusionMRI/tractseg_data/784565/data.nii.gz"
data = nib.load(pth).get_data()
```

```python
imgs = [data[72, :, :, 0].T, data[72, :, :, 1].T, data[72, :, :, 2].T, data[72, :, :, 3].T,
        data[:, 87, :, 0].T, data[:, 87, :, 1].T, data[:, 87, :, 2].T, data[:, 87, :, 3].T,
        data[:, :, 72, 0].T, data[:, :, 72, 1].T, data[:, :, 72, 2].T, data[:, :, 72, 3].T]
titles = [r"$b=0 s/mm^2$", r"$b=1000 s/mm^2$", r"$b=2000 s/mm^2$", r"$b=3000 s/mm^2$"]
```

```python
fig, axes = plt.subplots(3, 4, figsize=(20, 10))
for i, ax in enumerate(axes.flat):
    pcm = ax.imshow(imgs[i], cmap="gray", origin="lower")
    ax.set_title(titles[i%4])
    ax.set_axis_off()
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.05)
    fig.colorbar(pcm, cax=cax)
    
    if i==0:
        ax.set_ylabel("Sagittal")
    elif i==4:
        ax.set_ylabel("Coronal")
    elif i==8:
        ax.set_ylabel("Axial")
# plt.colorbar()
```
