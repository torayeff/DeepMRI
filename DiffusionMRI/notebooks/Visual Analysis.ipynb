# Visual Analysis

```python
import sys
from os.path import join
import numpy as np
import nibabel as nib
import sklearn.metrics
import matplotlib
from matplotlib import pyplot as plt

sys.path.append('/home/agajan/DeepMRI')
from deepmri import dsutils, visutils  # noqa: E402

%matplotlib inline
%load_ext autoreload
%autoreload 2
```

## Settings

```python
SUBJ_ID = "784565"
DATA_DIR = "/home/agajan/experiment_DiffusionMRI/tractseg_data/"
TRACT_MASKS_PTH = join(DATA_DIR, SUBJ_ID, "tract_masks", "tract_masks.nii.gz")
FEATURES_NAMES = ["PCA_COORDS", "MSCONVAE_COORDS"]
LABELS = ["Other", "CG", "CST", "FX", "CC"]
YLABELS = ["Ground Truth", "PCA (k=10)", "MS-AE"]
```

## Load Data

```python
dMRI = nib.load(join(DATA_DIR, SUBJ_ID, "data.nii.gz")).get_data()
```

```python
TRACT_MASKS = nib.load(TRACT_MASKS_PTH).get_data()
TRACT_MASKS = TRACT_MASKS[:, :, :, 1:]  # remove background class

pred_masks_list = []
for fn in FEATURES_NAMES:
    pred_masks_pth = join(DATA_DIR, SUBJ_ID, "pred_masks", fn + "_pred_masks.nii.gz")
    pred_masks_list.append(nib.load(pred_masks_pth).get_data())
```

## Visual Analysis

### Class: CG (Cingulum)
This class is best visible in Axial orientation

```python
tract_name = "CG"
orientation = "Coronal"
visutils.visualize_preds(dMRI, TRACT_MASKS, pred_masks_list,
                         LABELS, tract_name, orientation, 
                         YLABELS, slice_idxs=None, n_rows=3)
```

### Class: CST (Corticospinal tract)

```python
tract_name = "CST"
orientation = "Coronal"
visutils.visualize_preds(dMRI, TRACT_MASKS, pred_masks_list,
                         LABELS, tract_name, orientation, 
                         YLABELS, slice_idxs=None, n_rows=3)
```
