```python
import pandas as pd
import os
```

```python
df = pd.read_csv("adhd200_preprocessed_phenotypics.tsv", sep="\t")
df.head()
```

```python
train_ids = []
test_ids = []
all_ids = []

train_files = os.listdir('/home/agajan/data/')
test_files = os.listdir('/home/agajan/test_data/')

for img_name in train_files:
    idx = img_name.find('_session')
    scan_id = int(img_name[8:idx])
    train_ids.append(scan_id)
    all_ids.append(scan_id)

train_ids = list(set(train_ids))

for img_name in test_files:
    idx = img_name.find('_session')
    scan_id = int(img_name[8:idx])
    test_ids.append(scan_id)
    all_ids.append(scan_id)

test_ids = list(set(test_ids))
all_ids = list(set(all_ids))

print("Train: {}, Test: {}, All: {}".format(len(train_ids), len(test_ids), len(all_ids)))
```

```python
train_dict = {}
for sid in train_ids:
    dx = int(df[df['ScanDir ID'] == sid]['DX'].item())
    train_dict[sid] = dx
```

```python
test_dict = {}
for sid in test_ids:
    dx = df[df['ScanDir ID'] == sid]['DX'].item()
    if dx != 'pending':
        test_dict[sid] = int(dx)
```

```python
with open("train_adhd.csv", "w") as f:
    f.write("fmri,dx\n")
    
    for img_name in train_files:
        idx = img_name.find('_session')
        scan_id = int(img_name[8:idx])
        dx = train_dict[scan_id]
        
        f.write(img_name + "," + str(dx) + "\n")
```
